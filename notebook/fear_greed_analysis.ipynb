```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
sp500 = yf.download("^GSPC", start="2020-01-01", end="2024-01-01")

sp500 = sp500.reset_index()

sp500.head()
```

```python
data = pd.DataFrame()

data["Date"] = sp500["Date"]

data["Momentum"] = sp500["Close"].pct_change(20)

data["Volatility"] = sp500["Close"].rolling(20).std()

data["Strength"] = np.where(
    sp500["Close"] > sp500["Close"].rolling(50).mean(),
    1,
    0
)

data.head()
```

```python
np.random.seed(42)

data["Safe_Haven"] = np.random.normal(0,0.01,len(data))

data["Put_Call"] = np.random.uniform(0.8,1.2,len(data))

data["Breadth"] = np.random.uniform(0,1,len(data))

data.head()
```

```python
data = data.fillna(method="bfill")

data.shape
```

```python
data["Fear_Greed_Index"] = (
    data["Momentum"]*25
    - data["Volatility"]*25
    + data["Strength"]*15
    - data["Safe_Haven"]*10
    - data["Put_Call"]*10
    + data["Breadth"]*15
)
```

```python
data["Fear_Greed_Index"] = (
    (data["Fear_Greed_Index"] - data["Fear_Greed_Index"].min())
    /
    (data["Fear_Greed_Index"].max() - data["Fear_Greed_Index"].min())
)*100

data.shape
```

```python
data.to_csv("../data/fear_greed_index.csv", index=False)

print("CSV saved successfully")
```
